### Question:  10.How can you tell the difference between break and continue?

**Answer:**

break and continue are loop control statements executed inside a loop(with both loops for , while)

**break : Terminate** 

**continue: skip**

Difference : break stops the loop entirely, while continue skips the remaining code for the current iteration and moves on to the next iteration.

---------------------------------------------------------------------------------------------------------------------------
`When a code encounters loop control statement(break or continue) :`

* `"break" statement:` "break" statement terminates/stops the current loop completely and moves on to the next line of code outside the loop.It breaks the iteration.

* `"continue" statement:` "continue" statement skips the remaining code within the current iteration of the loop ,and moves on to the next iteration, if any.It skips the Iteration.





**Example of break statment:**

```python
for i in range(11):
    
    if i==8:
        print("Exiting!!!As i encountered Break!")
        break      
                   #As break statment executed, this terminates the loop completely at i is 8 and moves to next line,
                   #out of loop.
        print("Nothing more from this loop will get excuted,once break gets executed.")
                  #Note that ,This last print didn't get executed as break was before it.
            
    print(i)  #Gets executed for i= 0,1,2,3,4,5,6,7  
        
print("------------------------------------------")
print("Hello i came out of the loop Finally.")
```

### Question: 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

**Answer:**


All of these range(10),range(0,10) and range(0,10,1) will give same sequence of data.The only difference is the way arugments were specified here.

`range(10)`: 
* Here only stop value is there as 10(stop value not included).
* Interpreter took the Default start value as 0 and step_size as 1.
          
`range(0,10)`: 
* Here start value and stop value are there as 0 and 10 respectively
             (start value included , stop value not included).
* Interpreter took the  step_size as 1. 

`range(0,10,1)`: 
* Here start value , stop value and step size are there as 0 ,10,1 respectively
             (start value included , stop value not included).

```python
for i in range(10):
    print(i,end=' ')
```

```python
for i in range(0,10):
    print(i,end=' ')
```

```python
for i in range(0,10,1):
    print(i,end=' ')
```
