### Q1:What   is   the   role   of   the   'else'   block   in   a   try-except   statement?   Provide   an   example scenario  where  it  would  be  useful.
**Answer:**

* In a try-except statement, the 'else' block is an optional block and it provides a section of code to execute when no exceptions are raised within the 'try' block. 

* It is executed only if the 'try' block runs successfully without any exceptions being raised.

```python
def divide_numbers(x,y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    else:
        print("The result of the division is:", result)
```

```python
divide_numbers(100,2)
```

```python
divide_numbers(300,0)
```

### Q2.      Can   a   try-except   block   be   nested   inside   another   try-except   block?   Explain   with   an example.
**Answer:**

Yes we can use nested try except blocks in python.

Example:

```python
def nested_te_1():
    try:
        # Outer try block
        num1 = int(input("Enter a dividend: "))
        num2 = int(input("Enter a divisor: "))

        try:
            # Inner try block
            result = num1 / num2
            print("Result:", result)

        except ZeroDivisionError:
            print("Error: Division by zero is not allowed.")

    except ValueError:
        print("Error: Please enter valid integer values.")
```

```python
#Calling the function and passing input as 100 and 0 : Executes the inner try except block
nested_te_1()
```

```python
#Calling the function and passing the input as 100 and in string i will pass 'two' : Executes the outer try except blokck
nested_te_1()
```

```python
#Correct inputs.
nested_te_1()
```

**When we enter both the numbers:**

    - If both the inputs are correct means  both are non zero integer then Inner try block gets executed successfully.
    - If value as numerator is integer but denominator is zero (which is also as integer) then this through ZeroDivisionError and inner except block gets executed.
    - If value as numerator is integer but the denominator is not interger then outer except block will be executed as the value entered not correct.

### Q3.      How   can   you   create   a   custom   exception   class  in  Python?  Provide  an  example  that demonstrates  its  usage.
**Answer:**

```python
class CustomException(Exception):
    def __init__(self, message):
        self.message = message
```

```python
def div_numbers():
    try:
        num1 = int(input("Enter the value for numerator: "))
        num2 = int(input("Enter the value for denominator: "))

        if num2 == 0:
            raise CustomException("Division by zero is not allowed.")

        result = num1 / num2
        return result

    except ValueError:
        print("Invalid input. Please enter integers.")
    except CustomException as e:
        print("An error occurred:", e.message)
```

```python
div_numbers()
```

### Q4.      What  are  some  common  exceptions  that  are  built-in  to  Python?
**Answer:**

The common exceptions that are built in to Python :

* `TypeError`: Raised when an operation or function is performed on an object is not of appropriate type.

* `ValueError`: Raised when a function receives an argument of the correct type but an invalid value.

* `NameError`: Raised when a local or global name is not found.

* `IndexError`: Raised when a sequence subscript is out of range.

* `KeyError`: Raised when a dictionary key is not found.

* `FileNotFoundError`: Raised when a file or directory is requested but cannot be found.

* `ZeroDivisionError`: Raised when division or modulo operation is performed with zero as the denominator.

* `AttributeError`: Raised when an attribute reference or assignment fails.

* `ImportError`: Raised when an import statement fails to find the module definition or named attribute.

* `TypeError`: Raised when an operation or function is applied to an object of an inappropriate type.

* `KeyboardInterrupt`: Raised when the user interrupts the execution (typically by pressing Ctrl+C).

* `MemoryError`: Raised when an operation fails to allocate memory.

* `OverflowError`: Raised when the result of an arithmetic operation is too large to be expressed in the available numeric type.

### Q7.      What    are    log    formatters    in    Python    logging,    and   how   can   you   customise   the   log message  format  using  formatters?
**Answer:**

The log formatters in Python logging are object that define the strucutre like time, message , log level etc.

This allows the developer to customize the appearence and information included in log message.

By creating an instance of the "Formatter" class and specifying a format string with placeholders, developers can control the format of log messages.

The format string can include placeholders for timestamp, logger name, log level, and message content. 

```python
import logging 

logging.basicConfig(level=logging.INFO)

#Creating a logger.
logger_11=logging.getLogger("New_Assignment_11_logger")

handler_11=logging.FileHandler("New_Assignment_11_logger.log")
# Create a formatter
formatter = logging.Formatter('%(asctime)s  - %(name)s - %(levelname)s - %(message)s')

# Set the formatter for the file handler
handler_11.setFormatter(formatter)
logger_11.addHandler(handler_11)

def addition(a, b):
    logger_11.info("This is the start of the function")
    logger_11.info("The Elements are: %s, %s", str(a), str(b))
    return a + b
```

```python
addition(10,20)
```

```python
addition(100,200)
```

```python
addition(140,240)
```

**Answer:**

Print statements are useful for quick display ,during development because they allow us to display immediate information on the console. They are handy for quickly checking variable values or understanding the flow of our code.

Whereas 

Logging is better suited for long-term maintenance and troubleshooting in production environments. 
Logging provides structured and customizable log messages with severity levels, making it easier to track issues and analyze the behavior of our application over time.

### Q 10. Write    a    Python    program    that   logs   a   message   to   a   file   named   "app.log"   with   the following  requirements:

    ● The  log  message  should  be  "Hello,  World!" 
    ● The  log  level  should  be  set  to  "INFO."
    ● The  log  file  should  append  new  log  entries  without  overwriting  previous  ones.
**Answer:**

```python

import logging


logging.basicConfig(level=logging.INFO)             #log  level  should  be  set  to  "INFO."

logger_a=logging.getLogger("app logger")

#file named "app.log" .The logging FileHandler class appends log entries by default when it opens the log file.
handler_a=logging.FileHandler("app.log")   

# Create a formatter
formatter = logging.Formatter('%(asctime)s  - %(name)s  - %(levelname)s - %(message)s')

handler_a.setFormatter(formatter)
logger_a.addHandler(handler_a)

logger_a.info("Hello,  World!")                     #The  log  message  should  be  "Hello,  World!" 
```

### Q11. Create a Python program that logs an error message to the console and a file named "errors.log" if an exception occurs during the program's execution. The error message should include the exception type and a timestamp.
**Answer:**

**Note** 
       
       - FileHandler creates the file and addHandler adds message to it. This messages gets stored in file.
       - Streamhandler  with help of addHandler shows the messages into the console.

```python
import logging
```

```python
import logging

# Configure the logger
logging.basicConfig(level=logging.INFO)

# Create a logger
log_Q11 = logging.getLogger("errors logger")

# Create a formatter
formatter = logging.Formatter('%(asctime)s - %(name)s  - %(levelname)s - %(message)s')

# Create a file handler for the log file
file_handler_Q11 = logging.FileHandler("errors.log")
file_handler_Q11.setFormatter(formatter)

# Create a stream handler for console output
stream_handler_Q11 = logging.StreamHandler()
stream_handler_Q11.setFormatter(formatter)

# Add the handlers to the logger
log_Q11.addHandler(file_handler_Q11)
log_Q11.addHandler(stream_handler_Q11)

def division(num, deno):
    try:
        result = num / deno
        return result
    except ZeroDivisionError as e:
        
        log_Q11.error("Exception occurred: %s", e)
```
